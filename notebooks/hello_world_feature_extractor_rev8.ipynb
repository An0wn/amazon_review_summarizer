```python
# https://www.amazon.com/Tuft-Needle-TN23Q-Mattress-Queen/dp/B00QBZ265U/ref=sr_1_3?s=furniture&ie=UTF8&qid=1470623433&sr=1-3-spons&keywords=mattress&psc=1

doc1 = Loader('https://www.amazon.com/Tuft-Needle-TN23Q-Mattress-Queen/dp/B00QBZ265U/ref=sr_1_3?s=furniture&ie=UTF8&qid=1470623433&sr=1-3-spons&keywords=mattress&psc=1')
doc1.scrape(300)
doc1.extract(doc1.asin)
```

```python
# https://www.amazon.com/Intex-Pillow-Raised-Airbed-Electric/dp/B000HBILB2/ref=sr_1_3?ie=UTF8&qid=1470623449&sr=8-3&keywords=air+bed

doc2 = Loader('https://www.amazon.com/Intex-Pillow-Raised-Airbed-Electric/dp/B000HBILB2/ref=sr_1_3?ie=UTF8&qid=1470623449&sr=8-3&keywords=air+bed')
doc2.scrape(300)
doc2.extract(doc2.asin)
```

```python
corpus1 = ReviewSents(doc1)

unigramer1 = Unigramer()
unigramer1.candidate_unigrams(corpus1)

bigramer1 = Bigramer(unigramer1)
bigramer1.candidate_bigrams(corpus1)

trigramer1 = Trigramer(bigramer1)
trigramer1.candidate_trigrams(corpus1)

bigramer1.pop_bigrams(trigramer1)
unigramer1.update_review_count(bigramer1, trigramer1)
```

```python
corpus2 = ReviewSents(doc2)

unigramer2 = Unigramer()
unigramer2.candidate_unigrams(corpus2)

bigramer2 = Bigramer(unigramer2)
bigramer2.candidate_bigrams(corpus2)

trigramer2 = Trigramer(bigramer2)
trigramer2.candidate_trigrams(corpus2)

bigramer2.pop_bigrams(trigramer2)
unigramer2.update_review_count(bigramer2, trigramer2)
```

```python
polarizer1 = Polarizer(unigramer1, bigramer1, trigramer1)
polarizer1.polarize_aspects(corpus1)
```

```python
polarizer2 = Polarizer(unigramer2, bigramer2, trigramer2)
polarizer2.polarize_aspects(corpus2)
```

```python
match_data, comm_aspects = common_features(polarizer1, polarizer2)
```

```python
print_aspect_summary(comm_aspects, polarizer1, polarizer2, line_len=100)
```
